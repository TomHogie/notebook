```python
import numpy as np
import pandas as pd
import matplotlib
```

```python
file ='DATA\dictionary.csv'
countries_df=pd.read_csv(file)
countries_df.head()
```

```python
file ='DATA\summer.csv'
summer_df=pd.read_csv(file)
summer_df.head()

```

```python
summer_df.rename(columns={"Country":"Code"},inplace = True)
summer_df.head()
```

```python
summer_df.head()
```

```python
file ='DATA\winter.csv'
winter_df=pd.read_csv(file)
winter_df.rename(columns={"Country":"Code"},inplace = True)
winter_df.head()

```

```python
summer_countries_df= countries_df.merge(summer_df,how='inner',on='Code')
summer_countries_df["Season"] = "Summer"
```

```python
summer_countries_df.head()
```

```python
winter_countries_df= countries_df.merge(winter_df,how='inner',on='Code')
winter_countries_df["Season"] = "Winter"
winter_countries_df.head()
```

```python
all_df = pd.concat([summer_countries_df, winter_countries_df], sort=False)
all_df.head()
```

```python

percountry_df = all_df[all_df["Year"]>1984].groupby("Country").count()["Medal"].sort_values(by= "Medal", ascending=False)
```

```python
all_count_df = all_df[all_df["Year"] >= 1984] \
    .groupby(["Country"]) \
    .count()[["Medal"]] \
    .sort_values(by="Medal", ascending=False)
all_count_df.head(10)
```
